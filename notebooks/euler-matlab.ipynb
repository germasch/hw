After some sweat and tears, here's an equivalent piece of matlab code for the
Euler method / Skateboard problem.

It's not pretty, but at least it works. It's still worth it to look at the Python notebook for
additional information, as this is just a pretty basic implementation, by someone who doesn't
really know matlab well...

```octave
t_final = 20;
phi0 = .5;
y0 = [0, phi0];
M = 200;
YY = euler(@f, t_final, y0, M);
plot(YY(:,1), YY(:,2), YY(:,1), YY(:,3))

function E = euler(f, t_final, y0, M)
    dt = t_final / M;
    Y = zeros(size(y0, 2), M+1);
    T = 0:dt:t_final;
    Y(:,1) = y0;
    for j = 1:M
        Y(:,j+1) = Y(:,j) + dt * f(T(j), Y(:,j));
    end
    E = [T' Y'];
end

function [res] = f(t, y)
    g = 9.8;
    R = 5;
    om0 = g / R;

    omega = y(1);
    phi = y(2);
    res = [-om0^2 * phi; omega];
end
```
